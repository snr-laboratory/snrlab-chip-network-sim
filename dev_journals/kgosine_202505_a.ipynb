## Design Specifications

Specifications:
Penn's C-gain channel (no -3x amplifier): 25 uW per channel
Penn's Standard gain channel: 45 uW per channel

Speed:
Time to complete a reset cycle: 280ns (@ quoted power spec)

Clock speed:
The current clock is at 50MHz. From Texas A&M's simulation work with a reset scheme, the soft minimum seems to be 30MHz (such that the position resolution on reconstruction is not dominated by clock speed).

Requirements for this design:
Power under 45 uW per channel (ideally achieving 20 uW) with 50uW per channel being the maximum for 4mm pixel pitch.

Removing the 3x amplifier and making the comparator work much more efficiently are musts. We should design a dynamic comparator which is only on during clock edges. Then the replenishment logic will be even simpler since the comparaed logic is already aligned with the clock-edges.

For the replenishment current we only need a well controlled transistor.

Comparison to LarPix:
Their analog power consumption is 24 uW per channel.

Replenishment Current:
In Penn's design you can tune the replenishment charge between 0.3fC and 2fC and the clock rate.

Side points:
Penn used Vdd=1.6V (rather than 1.8V so that the transistors last longer in the cold).
