While the overall injection feedback behavior is preserved, there is obviously kickback from the clock through the latched comparator back to the integrator output and also at the comparator output, causing those voltage spikes.

When returning to this the priority is to minimize that feedback and also the voltage offset which is causing the comparator to fire at Vin=int_out= only 894mV rather than the 905mV which the Vref_comp is currently set to.

The large offset voltage is seen here:






Where the y-axis is the comparator output and the x-axis is Vin (Vref=900mV).
