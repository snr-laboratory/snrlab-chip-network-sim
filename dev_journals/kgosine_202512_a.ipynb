### Monte Carlo Simulation


Variant: non-shared resistor (transistor-resistor)

1000 samples with both mismatch & process variations. Measured the following:

(1) The time of the first Q pulse (true call for replenishment)

(2) The time of the first OUTP pulse (call for replenishment exiting chip)

(3) Integral of the integrator output

(4) Number of Q pulses

(5) Number of OUTP pulses

(6) Average replenishment charge (charge replenished/number of Q pulses)

(7) Charge replenished

All in response to a 2fC input charge signal.

![](https://drive.google.com/uc?export=view&id=1fG4lFY8rZGy6m_AcBaY2UxTE-W_LOa-l)
![](https://drive.google.com/uc?export=view&id=1Oquov_dpL6OJnkPB4SM6puFQaVDg9OY2)

Overall good, a few events with 5 replenishments but the vast majority had 4. For all the replenishment charge amount was consistently half a fC as expected.

Variant: shared resistor with feedback resistor. Measured the same outputs.
The main difference is the average replenishment size is larger, at 0.815fC.
