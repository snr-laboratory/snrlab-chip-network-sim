## Sizing Clocked Comparator


Spec goals:
*   <10uW average power consumption
*   Able to drive a 0.9V swing across 50fF load cap within 2ns

Design steps:
Get $g_m$ from timing constraint
Select inversion level
Get $I_d$
Get W

At the time of evaluation (Clk rising edge), the latch behaves as a linear positive feedback amplifier, with the differential voltage at the two internal latch nodes grows exponentially:
$$ \Delta v(t) = \delta V_0 e^{t/ \tau}$$
Where $ \tau = \frac{C_{node}}{g_{m, eff}}$

Solve $ \Delta v(t_{reg})= V_{swing}$ for $g_{m, eff}$

Drive constraint:
$$ t_{reg}= \frac{C_{node}}{g_{m,eff}}ln(\frac{V_{swing}}{\Delta V_0})$$
Constraint: $t_{reg}<3ns$

Where $V_{swing}$= 0.9V
$\Delta V_0$=5mV (input offset for switching)
$C_{node}$=50fF

$$g_{m, eff} = \frac{C_{node}}{t_{reg}}ln(\frac{V_{swing}}{\Delta V_0})$$

$g_{m, eff}= 0.121 mS$
Where $g_{m, eff}= g_{m, n}+g_{m, p}$

Where M0&M1 are the PMOS cross coupled latches and M12&M13 are the NMOS cross coupled latches. Let $g_{m, n}=g_{m, p}=\frac{1}{2}g_{m, eff}=60.5 uS$






#### Selecting $g_m/I_d$ & $I_d$

If $g_m/I_d$= 8 (moderate inversion) we get decently low current draw while not going into weak inversion which causes very large widths and therefore capacitance, slowing down the circuit.

$ I_d=\frac{g_m}{g_m/I_d} = \frac{60 \mu S}{8 V^{-1}}= 7.5\mu A$

#### Select length
For the cross coupled latching transistors, chose minimum L for speed.

$$W_{PMOS} = \frac{I_d}{I_d/W}= \frac{7.5uA}{12.45A/\mu m} = 602nm $$

For L=180nm and $g_m/I_d$= 8, $I_d/W_{NMOS}$= 37.58 A/um
$$W_{NMOS} = \frac{I_d}{I_d/W}= \frac{7.5uA}{37.85A/\mu m} = 200nm $$
