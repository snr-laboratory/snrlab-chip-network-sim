## Sizing OpAmp


Constraints:

*   Average power consumption less than 40 uW
*   Integrator should respond to a 50nA, 20ns (1fC) current pulse within 10ns

#### Response time:
For a 1fC input charge through a 50fF feedback cap, the integrator output should change by 20mV. This places a constraint on the closed loop frequency of the circuit.
For a single-stage circuit, the output grows as $ y(t) = 1 - e^{-t/\tau}$ such that the time to reach a fraction of the total output is $t_p=-\tau ln(1-p)$ where $p=0.9$ (to reach 90% of the output). Therefore, the required time constant is given by:
$$\tau = -\frac{t_{90}}{ln(1-0.9)}= -\frac{10ns}{2.3}\approx 4.323ns $$
And for a single pole system,
$$ f_{CL} \approx \frac{1}{2 \pi \tau}= 36.64 MHz$$

The total capacitance at the output node is the feedback cap + load capacitance. This is related to the closed loop frequency by:
$$ f_{CL} \approx \frac{g_{m, eff}}{2 \pi C_{node}} $$
Therefore,
$$ g_{m, eff} = 2 \pi f_{CL}C_{node}= 2 \pi (36.64 MHz)(550fF) = 126 \mu S$$

Where $g_{m, eff}$ is the transconductance of the input pair.

#### Power consumption
$$ I = \frac{P}{V_{DD}}\leq \frac{40 \mu W }{1.8V}= 22.2 \mu A $$

#### Sizing input Pair
Because of the large transconductance requirement on the input pair, the majority of the current is allocated here to get a reasonable $g_m/I_d$ for PMOS. The problem here is that choosing a large $g_m/I_d$ will increase the widths of the input transistors and therefore the input capacitance of the opamp. This could prove to be a problem if the the electron signal can't drive this amount of input capacitance (?).
If the input capacitance of this circuit does prove to be an issue, there is no solution other than to add an output stage to the opamp. Because as stands, the speed response requirement places a hard requirement on $g_m/I_d$ and within the power constraint, no more current can be allocated to the input branch above this.

Target: $g_m = 126 \mu S$
$$I_d = 8 \mu A$$
$$ g_m/I_d = 15.75 $$

$$ L = 180nm$$
$$ I_d/W = 2.02 \mu A /\mu m$$
$$ W = \frac{I_d}{I_d/W}= \frac{8 \mu A }{2.02 \mu A /\mu m} = 3.96 \mu m $$

## Sizing Clocked Comparator


Spec goals:
*   <10uW average power consumption
*   Able to drive a 0.9V swing across 50fF load cap within 2ns

Design steps:
Get $g_m$ from timing constraint
Select inversion level
Get $I_d$
Get W

At the time of evaluation (Clk rising edge), the latch behaves as a linear positive feedback amplifier, with the differential voltage at the two internal latch nodes grows exponentially:
$$ \Delta v(t) = \delta V_0 e^{t/ \tau}$$
Where $ \tau = \frac{C_{node}}{g_{m, eff}}$

Solve $ \Delta v(t_{reg})= V_{swing}$ for $g_{m, eff}$

Drive constraint:
$$ t_{reg}= \frac{C_{node}}{g_{m,eff}}ln(\frac{V_{swing}}{\Delta V_0})$$
Constraint: $t_{reg}<3ns$

Where $V_{swing}$= 0.9V
$\Delta V_0$=5mV (input offset for switching)
$C_{node}$=50fF

$$g_{m, eff} = \frac{C_{node}}{t_{reg}}ln(\frac{V_{swing}}{\Delta V_0})$$

$g_{m, eff}= 0.121 mS$
Where $g_{m, eff}= g_{m, n}+g_{m, p}$

Where M0&M1 are the PMOS cross coupled latches and M12&M13 are the NMOS cross coupled latches. Let $g_{m, n}=g_{m, p}=\frac{1}{2}g_{m, eff}=60.5 uS$






#### Selecting $g_m/I_d$ & $I_d$

If $g_m/I_d$= 8 (moderate inversion) we get decently low current draw while not going into weak inversion which causes very large widths and therefore capacitance, slowing down the circuit.

$ I_d=\frac{g_m}{g_m/I_d} = \frac{60 \mu S}{8 V^{-1}}= 7.5\mu A$

#### Select length
For the cross coupled latching transistors, chose minimum L for speed.
