V1/PLUS is injection current from transmission gate structure, V14/PLUS is injection current from Qbarcontrolled structure.

For both structures, having a smaller resistive load to ground (1Meg) led to sharper changes in current during switch transistions. For very large resistors (10Meg), we see in the voltage plots that there is some RC behavior between the load resistor and the capacitance of the transistors in the structure (worse in the Qbarcontrolled structure).

With a smaller load resistor we see more square pulses (quicker transition times, less RC coupling) at the cost of large swings in current outside of the desired 50nA pulses. Overall, the transmission gate structure performed better.
