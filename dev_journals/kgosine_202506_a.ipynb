### Returning to charge injection: No ideal current source, instead resistor across constant voltage drop

Since the voltage that the injection current is seeing is at the input of the integrator is constant (virtual ground between Vref=900mV), we can use a transistor as a switch before an appropriately sized resistor. Since there is a voltage drop from 1.8V to Vref=0.9V and we want a current injection of 50nA:
$$ R = \frac{\Delta V }{I}= \frac{900mV}{50nA}=18M\Omega  $$

This is our best result in terms of a constant ON current of exactly 50nA and no current peaks at switching. There is some delay in turning off current but ultimately, the charge in this region is significantly less than the charge from when the injection is ON.

The primary challenge with this design is fitting an 18MegOhm resistor on our layout.
