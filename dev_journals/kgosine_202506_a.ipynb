### Cascode Current Mirror

The cascode current mirror has boosted output impedance which may stabilize the current injection. Used to replace the current mirror in the transmission gate configuration so it is actually a combination of transmission gate + cascode current mirror:



<img src="https://drive.google.com/uc?id=1GHX1t6yK_oFKw8tnHEYwgXxRLYTUPTwd"  height="500" >

V15/PLUS is the cascode current mirror structure injection current.

For the small load resistance (1Meg), the output current is much more stable in the high state in this configuration (compare red traces in strips 2&3) but the value of the output current is much larger (1.18uA) for this configuration compared to just the transmission gate scheme (175na).

Comparison of the large load resistance (10Meg) between the transmission gate and the cascode current mirror structures are shown in the bottom strip and seem to be quite similar. In the transmission trace however, in each pulse the maximum current slowly reduces from 107nA in the first pulse to 81nA in the 7th pulse while with the cascode current mirror, the high current is a constant 116nA.

Both structures still suffer from the same large time constant for high load resistances.

It seems the cascode current mirror improves the stability of the output current although doesn't address the overshoot in current or the time constant for turning off the current.


### Injecting current into Vref instead of Vss

This limits the output voltage swing (unclear why the voltage goes up to about 1.2V) since it now goes from 900mV to about 1.2V and not from 0V. This worked well in reducing the output current for all resistive loads down to the order of hundreds of nA.
