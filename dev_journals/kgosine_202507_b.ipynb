### Output Stage Options

Choice between a common source stage and a source-follower (commmon drain) stage.

Common source: ie class A output stage, provides voltage gain, inverting, downside is it has a large output impedance.

Source follower: gain is less than one, non-inverting, low output impedance (good current driver), introduces Vgs drop between input and output, limits output swing (source cannot reach gate voltage and stay on while the drain voltage may equal the gate voltage and remain in saturation)

### Op-Amp Topology Options

Folded cascode has a higher power draw because there is an extra set of transistors to bias. The telescopic cascode is simpler but stacking of transistors between Vss and Vdd reduces the output voltage swing. I'm thinking that since our output voltage range should not be very large (replenishments always try to keep it near baseline) and is only in one direction (integrator output should only ever go up, we don't need it to swing down), we can lower Vref, the integrator baseline, closer to Vss and allow more of the voltage swing to be accessible.
