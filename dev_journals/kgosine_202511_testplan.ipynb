### Concerns & Notes

Voltage biasing the pixel pad? With a DAC?

Channels have their own clk reciever. This means each channel needs its own pad for RX biasing current. Channels do all get the same LVDS clock input.

Similarly, they don't share an opamp biasing circuit which means each channel needs 1 pad for the bias ctrl current.

Keeping these separate may make sense for prototyping but not if we don't want that many pads.

## Application of Bias Currents

SPY_CURR[0:2]: 1.25V across 10K for 3.7uA per channel
RX_CURR[0:2]: 1.15V across 10K for 5.4uA per channel
BIAS_CTRL[0:2]: 0V across 200K for 2.65uA per channel

## Description of Variants

(1) **feedback_resistor**

NMOS transistor with gate connected to pad (nominally connected to VSS) in parallel with the CSA feedback capacitor. replenishment circuit is resistor-transistor with shared resistor across variants. CH0

(2) **noresistor**

Only feedback capacitor across CSA in/outputs. Assumes DC voltages set by QPix feedback loop. replenishment circuit is resistor-transistor with shared resistor across variants. CH1

(3) **fbres_transistor_resistor**

NMOS transistor with gate connected to pad (nominally connected to VSS) in parallel with the CSA feedback capacitor + replenishment circuit with localized resistor (transistor followed by resistor connected to CSA input). CH2
