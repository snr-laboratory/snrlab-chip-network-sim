### Sizing Input PMOS Pair on Folded Cascode

<img src="https://drive.google.com/uc?id=1JskvXOpxF27Xa4DSfRDDpH_dQ1Eqntcb"  height="300" >

Our primary constraint is power within which we want to optimize for gain.
Given Vdd=1.8V, and we want the opamp to (at absolute maximum) to consume 20 uW of power. Therefore, the total current draw for the circuit should be about 11uA.
$$ I_{DD}= \frac{P}{V_{DD}}= \frac{20 \mu W}{1.6V}=11.1 \mu A $$

I somewhat arbitrarily decided to alocate half the current to the differential leg and the load leg.

M13 makes up the tail current of the differential pair. Letting $I_{tail}=5 \mu A$, the drain current through each of the input PMOS is $2.5 \mu A$.

Because the differential stage is the gain stage of this circuit (this is a one-stage folded-cascode, no output stage, the gm/Id was chosen to be in the moderate inversion region (between 15-20) which balances efficiency and speed. Because PMOS transistors are less efficient than NMOS transitors, gm/Id is maxed out at 20. I chose gm/Id=16. Given this and our allocated drain current, we can calculate the gm of the transistor:
$$ g_m = (\frac{g_m}{I_d})(I_d)=16 V^{-1} \cdot 2.5 \mu A= 40 \mu S$$

In the gm,ro versus gm/Id graph, at gm/Id=16, I selected a length and noted the corresponding intrinsic gain. I selected length=700nm. Longer lengths would have corresponded to a higher intrinsic gain but gm,ro= 189 seemed sufficient for a first pass.

I determined the width from the Id/W versus gm/Id plot. On the L=700nm curve I found the Id/W corresponding to gm/Id=16 was $312\cdot 10^{-3} A/m$ and since we know Id, I can calculate the width:
$$ W = \frac{I_d}{\frac{I_d}{W}}= \frac{2.5 \mu A}{16 V^{-1}}=8 \mu m$$


I applied these W & L to the input PMOS transistors. Adjusted Ibias3 which is the tail current of the diff pair. Also adjusted Ibias1 which acts as the current source of the loads which was set to 3uA (3uA on two legs for a total of 6uA in the load leg). Confirmed that all the transistors were still in saturation and the gm and gmro of these two transistors where what was expected from the calculations.
