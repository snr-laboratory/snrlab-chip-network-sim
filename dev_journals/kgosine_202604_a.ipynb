## Hydra Network

hydra_ctrl.sv receives local packets from the even_router.sv and packet traffic from comms_ctrl.sv

These two data sources meet at the priority_fifo_arbiter.sv where config/readback packets take priority over event packets should they arrive at the same time.

Hydra will tell upstream logic if it can accept another packet such that event_router.sv will only grab from a local channel FIFO when Hydra is ready. The selected packet (from the fifo arbiter) is written to the shared FIFO (fifo_latch.sv).

Hydra waits until a downstream UART is available and reads a packet from the shared FIFO. It loads that packet into the TX side of the enabled UART lane.

Hydra will broadcast the same packet (for tranmsission) to every enabled downstream UART line. The network configuration is what determines which UART line is enabled for each chip. Which UART is enabled is controlled by configuration register bits.
