## Placing Source Follower Outside of Loop



This should avoid the compensation and stability problems of the two stage op amp. This structure has a phase margin of only 35 degrees with no compensation:

<img src="https://drive.google.com/uc?id=1Uh4NdG5OaKALSJDE1CecOXFKyGLAmvo_"  height="300" >
<img src="https://drive.google.com/uc?id=1e511anDrmRwwC1PwKmejsEEKcMVtVAuz"  height="300" >

Adding a compensation capacitor for in the same place as before. A 230fF capacitor is the minimum needed to get a phase margin above 65 degrees. Recall that when the SF is inside the loop, a 500fF compensation capacitor only offers a 55 deg phase margin...

#### Integration Performance
With compensation (230fF for 65 deg phase margin), the rise time is reduced to about 45ns. The steps are no longer 8mV but a consistent 6.5mV (gain of source follower is only about 0.8).

<img src="https://drive.google.com/uc?id=1jXBtuo7hTkrxW2S2ccLWedei3dMXGqVq"  height="300" >


With no compensation, oscillations are not visible (35 deg phase margin) and the rise time is not much quicker, still around 40ns. The steps are 6.35mV.

<img src="https://drive.google.com/uc?id=18WRAKwmFUrcIQyphMiuQnzeJwnuL3ljx"  height="300" >
