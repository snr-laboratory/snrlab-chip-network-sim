#### Placing source follower buffer outside of loop
This should avoid the compensation and stability problems of the two stage op amp. This structure has a phase margin of only 35 degrees with no compensation:

#### Integration Performance
With compensation, the rise time is reduced to about 45ns. The steps are no longer 8mV but a consistent 6.5mV (gain of source follower is only about 0.8).
